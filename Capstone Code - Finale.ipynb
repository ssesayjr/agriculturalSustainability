# Capstone: Agricultural Sustainability in West Africa 

# Introduction

The initial question posed for the Analytic report and Research Proposal Capstone is to determine if smallholder agricultural farmers, businesses, and entrepreneurships are lucrative and sustainable in Sierra Leone, Senegal, Guinea-Bisaau and/or Guinea. The question stems from US Government Agencies, International Development sector, and Non-Government Organizations redirecting and focusing funds into the agricultural sector of developing countries. There is the logical assumption that developing countries need a sustainable and growing agricultural market; however, in the United States, 2 percent of the population are farmers but feed upwards to 330 million people annually (source: American Farm Bureau Federation).  Based on this assumption, a smaller percentage of farmers would be able to sustain either of the four countries. By analyzing the growth of the agricultural sector in four Western African countries and the effect on the country's Gross Domestic Product this capstone will determine which West African Country would be best for an agricultural smallholder farmer, business or entrepreneurs.

```python
## Importing Numpy/MatPlotLib and Pandas
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

# Data Set 

Several different datasets were used for the analytics of this report hosted/sponsored by government organizations and non-government entities. The following data sources utilized were: 

- American Farm Bureau Federation - an independent, non-governmental, voluntary organization governed by and representing farm and ranch families; united for the purpose of analyzing their problems and formulating action to achieve educational improvement, economic opportunity and social advancement, thereby promoting the national well-being. 
- Economic Community of West African States - a 15-member regional group representing Benin, Burkina Faso, Cape Verde, Côte d’ Ivoire, The Gambia, Ghana, Guinea, Guinea-Bissau, Liberia, Mali, Niger, Nigeria, Sierra Leone, Senegal and Togo with a mandate of promoting economic integration in all fields of activity in the constituting countries. 
- African Development Bank Group Data Portal 
- United Nations Population Division


```python
## Pandas module reading in the country GDP files 
sierra_leone = pd.read_csv('Sierra Leone_T.csv')
guinea_bissau = pd.read_csv('Guinea-Bissau_T.csv')
guinea = pd.read_csv('Guinea_T.csv')
senegal = pd.read_csv('Senegal_T.csv')

## Pandas module reading in West African exports setting export years
country_Export = pd.read_csv('West Africa Export to US (dollars).csv')
## Set dataframe for the available years provided by the GDP document 
years = pd.DataFrame([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016 ])

## Create dataframe for the years available in export dataframe 
export_Year = pd.DataFrame([2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016 ])
```

```python
## Create line graph with country data points 
plt.plot(years,sierra_leone.iloc[0:,2],  color='green', label='Sierra Leone')
plt.plot(years,guinea_bissau.iloc[0:,2],  color='purple', label='Guinea Bissau')
plt.plot(years,guinea.iloc[0:,2],  color='red', label='Guinea')
plt.plot(years,senegal.iloc[0:,2],  color='yellow', label='Senegal')

## Setting the graph titles and axises 
plt.title('Percentage of Available Agricultural Land By Country')
plt.xlabel('Years')
plt.ylabel('Values in %')
plt.legend(loc='best')

## Shows the plot of 4 countries agircultural land 
plt.show()
```

```python
## Reads the agricultural trade data from each country export data frames 
plt.plot(country_Export.iloc[0: ,0], country_Export.iloc[0: ,1], label='Senegal', color ='purple' )
plt.plot(country_Export.iloc[0: ,0], country_Export.iloc[0: ,2], label='Guinea', color ='green' )
plt.plot(country_Export.iloc[0: ,0], country_Export.iloc[0: ,3], label='Guinea Bissau', color ='blue' )
plt.plot(country_Export.iloc[0: ,0], country_Export.iloc[0: ,4], label='Sierra Leone', color ='orange' )

## Setting the graph titles and axises
plt.title('Percentage of Agricultural Trade over Time by Country')
plt.xlabel('Years')
plt.ylabel('Percentage of GDP')
plt.legend(loc='best')

#Shows the plot of percentage of agricultural land
plt.show()
```

Figure 1A demonstrates that the highest percentage of the available agricultural land is in Guinea, Senegal having the lowest percentage with highest fluctuation, and Sierra Leone with roughly 65% agricultural at the end of 2015. 

The previous graph demonstrates the availability of agricultural land but doesn’t demonstrate if the land was profitable. To answer this question requires more information regarding the country’s GDP. Figure 2A looks at the Gross Domestic Product of Agricultural Trade, adding more insight.

Although the data shows a negative correlation between arable land and GDP of .16, the data raises some areas of concern. Primarily the correlation coefficient between arable land and agricultural land. Figure 1A shows that Guinea Bissau percentage of available agricultural lands is between 65%-70%, and the correlation coefficient further suggests that the same increase in agricultural land is exactly inline with an increase of arable land. We would expect to see a very close coefficient between the two, but not 1 due to natural disasters or conversion of arable for city development. Guinea Bissau has the lowest coefficient correlation among the four countries. 


```python
#Correlation Coefficient: adding GDP to dataframe and finding CC 
corr_guineaBis = guinea_bissau.join(country_Export.iloc[0:10,3])
corr_sierraL = sierra_leone.join(country_Export.iloc[0:10,4])
corr_guinea = guinea.join(country_Export.iloc[0:10,2])
corr_senegal = senegal.join(country_Export.iloc[0:10,1])

#Correlation Coefficient for Guinea Bissau  
corr_guineaBis.corr(method = 'pearson')
```
